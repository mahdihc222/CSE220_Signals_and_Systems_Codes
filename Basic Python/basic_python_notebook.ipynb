```python
class Animal:
    def __init__(self, name, age):
        self.name = name 
        self.age = age 
    def print_details(self):
        print('Name: {}, Age: {}'.format(self.name,self.age))
    def talks(self):
        return f'{self.name} talks'
    def __str__(self):
        return f'{self.name} and {self.age}'
```

```python
a = Animal(2,3)
a.print_details()
print(a)
```
